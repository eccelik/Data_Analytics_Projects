# KNN Regressor

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
fig,ax=plt.subplots(figsize=(7,7))
k_list=np.arange(1,30,1)
knn_dict={} # To store k and mse pairs
for i in k_list:
#Knn Model Creation
    knn=KNeighborsRegressor(n_neighbors=int(i))
    model_knn=knn.fit(X_train,y_train)
    y_knn_pred=model_knn.predict(X_test)
#Storing MSE 
    mse=mean_squared_error(y_test,y_knn_pred)
    knn_dict[i]=mse
#Plotting the results
ax.plot(knn_dict.keys(),knn_dict.values())
ax.set_xlabel('K-VALUE', fontsize=15)
ax.set_ylabel('MSE' ,fontsize=15)
ax.set_title('ELBOW PLOT' ,fontsize=25)
```

```python
knn_reg=KNeighborsRegressor().fit(X_train,y_train)
```

```python
# Display the score values by calling knn
print("Training set score: {:.2f}".format(knn_reg.score(X_train, y_train)))
print("Test set score: {:.2f}".format(knn_reg.score(X_test, y_test)))
```

Random forest turn out a great result on training but has overfitting even without hyper-parameter tuning but it had overfitting problem.

To overcome the overfitting problem we did hyper paremeters tunning and the the result got better(high on training and testing). 

The reason why Random Forest turn out the best result:

-The "forest" it builds is an ensemble of decision trees, usually trained with the “bagging” or "Bootstrap Aggregation" is a sampling technique in which we randomly sample with replacement from the data set.
-The general idea of the bagging method is that a combination of learning models increases the overall result.

-Random forest builds multiple decision trees and merges them together to get a more accurate and stable prediction.

-Instead of searching for the most important feature while splitting a node, it searches for the best feature among a random subset of features. This results in a wide diversity that generally results in a better model.
 
When bootstrapping, we use only about 2/3 of the data. The approximately 1/3 of the data (“out-of-bag” data) is not used in the model and can conveniently be used as a test set.

-Also Random forest handles outliers by essentially binning them. It is also indifferent to non-linear features.

The Random Forest is the best model out of 3 models.
